```python
"""
Aim

To perform data cleaning and feature encoding using Pandas and Scikit-learn for preprocessing raw datasets, improving data quality, and transforming categorical attributes into numerical format suitable for machine learning and data analysis applications.

Description

Data preprocessing is an essential step in the data science and machine learning workflow. Real-world datasets are often incomplete, inconsistent, noisy, and contain categorical values that cannot be directly processed by machine learning algorithms. This experiment aims to prepare such datasets for analysis by applying data cleaning and feature encoding techniques using Python libraries such as Pandas and Scikit-learn.

Data Cleaning using Pandas

In the first stage of the experiment, the dataset is loaded and examined for inconsistencies and missing information. Various preprocessing operations are performed, including:

Identifying and handling missing or null values using methods such as removal or replacement,
Removing duplicate records to improve data reliability,
Correcting inconsistent data entries and formatting issues,
Detecting and handling outliers that may affect model performance,
Renaming columns and converting data types where necessary,
Organizing and structuring the dataset for further processing.

These steps improve the overall quality, consistency, and accuracy of the dataset.

Feature Encoding using Scikit-learn

Many machine learning algorithms cannot process categorical or text-based data directly. Therefore, categorical features must be converted into numerical representations. In this stage, different encoding techniques provided by Scikit-learn are applied, such as:

Label Encoding – converts categorical labels into integer values,
One-Hot Encoding – transforms categories into binary vectors,
Ordinal Encoding – encodes ordered categories based on rank or hierarchy.

Feature encoding helps preserve meaningful information from categorical data while making the dataset compatible with machine learning models.
"""
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# Load dataset (change file name as needed)
df = pd.read_csv("raw_merged_heart_dataset.csv")

# Display first 5 rows
print("Original Data:")
print(df.head())

# -----------------------------
# 🔹 1. DATA CLEANING
# -----------------------------

# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values

# For numerical columns → fill with mean
num_imputer = SimpleImputer(strategy='mean')
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = num_imputer.fit_transform(df[num_cols])

# For categorical columns → fill with most frequent value
cat_imputer = SimpleImputer(strategy='most_frequent')
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])

# -----------------------------
# 🔹 2. FEATURE ENCODING
# -----------------------------

# Label Encoding (for ordinal or simple categories)
label_encoder = LabelEncoder()

for col in cat_cols:
    df[col] = label_encoder.fit_transform(df[col])

# OR use One-Hot Encoding (better for non-ordinal data)
# df = pd.get_dummies(df, columns=cat_cols)

# -----------------------------
# 🔹 FINAL OUTPUT
# -----------------------------

print("\nCleaned & Encoded Data:")
print(df.head())

# Save cleaned dataset
df.to_csv("cleaned_data.csv", index=False)
```
